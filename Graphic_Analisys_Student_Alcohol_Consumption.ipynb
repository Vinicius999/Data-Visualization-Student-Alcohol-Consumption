```python
def calcular_media(row):
  return round((row['G1'] + row['G2'] + row['G3']) / 3, 2)

df['média'] = df.apply(lambda row: calcular_media(row), axis=1)

df.head()
```

## **Gráficos com Matplotlib**
#### **Qual a proporção dos estudantes que querem fazer faculdade?**


```python
dataT = df['higher'].value_counts(normalize=True).values
dataM = df[df['sex'] == 'M']['higher'].value_counts(normalize=True)
dataF = df[df['sex'] == 'F']['higher'].value_counts(normalize=True)

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Pretendem fazer Faculdade', ha = 'center', va = 'center', fontsize = 20)

axs[0].set_title('Todos')
axs[0].pie(
  dataT,
  labels=['Sim', 'Não'],
  shadow=True,  
  wedgeprops={'linewidth': 1, 'edgecolor': "blue"},
  explode=(0, 0.1),             
  autopct='%1.1f%%')

axs[1].set_title('Homens')
axs[1].pie(
  dataM,
  labels=['Sim', 'Não'],
  shadow=True,  
  wedgeprops={'linewidth': 1, 'edgecolor': "blue"},
  explode=(0, 0.1),
  autopct='%1.1f%%')

axs[2].set_title('Mulheres')
axs[2].pie(
  dataF,
  labels=['Sim', 'Não'],
  shadow=True,
  wedgeprops={'linewidth': 1, 'edgecolor': "blue"},
  explode=(0, 0.1),
  autopct='%1.1f%%')

plt.show()
```

#### **Querer fazer faculdade influencia no desempenho dos estudantes?**

```python
dataS = df[df['higher'] == 'yes']['média']
dataN = df[df['higher'] == 'no']['média']

dataMY = df[df['higher'] == 'yes'][df['sex'] == 'M']['média']
dataMN = df[df['higher'] == 'no'][df['sex'] == 'M']['média']

dataFY = df[df['higher'] == 'yes'][df['sex'] == 'F']['média']
dataFN = df[df['higher'] == 'no'][df['sex'] == 'F']['média']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Influência do desejo de ir para Faculdade', ha = 'center', va = 'center', fontsize = 20)

axs[0].set_title('Todos')
axs[0].scatter(dataS.index, dataS, label='Desejam ir', color='blue')
axs[0].scatter(dataN.index, dataN, label='Não desejam ir', color='red')

axs[1].set_title('Homens')
axs[1].scatter(dataMY.index, dataMY, label='Desejam ir', color='blue')
axs[1].scatter(dataMN.index, dataMN, label='Não desejam ir', color='red')

axs[2].set_title('Mulheres')
axs[2].scatter(dataFY.index, dataFY, label='Desejam ir', color='blue')
axs[2].scatter(dataFN.index, dataFN, label='Não desejam ir', color='red')

plt.legend()
plt.show()
```

#### **Aulas pagas de reforço influenciam no desempenho dos estudantes?**

```python
dataS = df[df['schoolsup'] == 'yes']['média']
dataN = df[df['schoolsup'] == 'no']['média']

dataMY = df[df['schoolsup'] == 'yes'][df['sex'] == 'M']['média']
dataMN = df[df['schoolsup'] == 'no'][df['sex'] == 'M']['média']

dataFY = df[df['schoolsup'] == 'yes'][df['sex'] == 'F']['média']
dataFN = df[df['schoolsup'] == 'no'][df['sex'] == 'F']['média']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Influência do Reforço', ha = 'center', va = 'center', fontsize = 20)

axs[0].set_title('Todos')
axs[0].scatter(dataS.index, dataS, label='Com reforço', color='red')
axs[0].scatter(dataN.index, dataN, label='Sem reforço', color='blue')

axs[1].set_title('Homens')
axs[1].scatter(dataMY.index, dataMY, label='Com reforço', color='red')
axs[1].scatter(dataMN.index, dataMN, label='Sem reforço', color='blue')

axs[2].set_title('Mulheres')
axs[2].scatter(dataFY.index, dataFY, label='Com reforço', color='red')
axs[2].scatter(dataFN.index, dataFN, label='Sem reforço', color='blue')

plt.legend()
plt.show()
```

## **Gráficos com Seaborn**

```python
df['traveltime'].replace(1, '<15 min.', inplace = True)
df['traveltime'].replace(2, '15 a 30 min.', inplace = True)
df['traveltime'].replace(3, '30 min. a 1 hora', inplace = True)
df['traveltime'].replace(4, '>1 hora', inplace = True)
df['Pstatus'].replace('T', 'Pais Juntos', inplace = True)
df['Pstatus'].replace('A', 'Pais Divorciados', inplace = True)
df['famrel'].replace(1, 'Muito Ruim', inplace = True)
df['famrel'].replace(2, 'Ruim', inplace = True)
df['famrel'].replace(3, 'Neutro', inplace = True)
df['famrel'].replace(4, 'Bom', inplace = True)
df['famrel'].replace(5, 'Excelente', inplace = True)
```

#### **O tempo de deslocamento de casa para a escola influencia nas notas?**

```python
df_traveltime1 = df.pivot_table(values = 'média', index = 'traveltime', aggfunc = 'mean').sort_values(by = 'média', ascending = False).reset_index()


sns.set(style='whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
fig.suptitle('Análise por Tempo de Deslocamento', ha = 'center', va = 'center', fontsize = 20)

# Tempo de Deslocamento
plt.subplot(1, 2, 1)
sns.histplot(df.traveltime, palette = 'Blues') 
plt.xlabel('Tempo de Deslocamento')
plt.title('Distribuição Tempo de Deslocamento', fontsize = 14)


# Tempo de Deslocamento por Nota Média Geral
y = df_traveltime1['média'].sort_values(ascending=False)
x = df_traveltime1['traveltime']
plt.subplot(1, 2, 2)
sns.barplot(x, y, palette = 'Blues')
plt.xlabel('Tempo de Deslocamento')
plt.ylabel('Média Geral')
plt.title('Tempo de Deslocamento por Nota Média Geral', fontsize = 14)


plt.show()
```

#### **O relacionamento dos pais influencia nas notas e no consumo de álcool?**

```python
sns.set(style='darkgrid')
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
fig.suptitle('Análise por Status dos Pais', ha = 'center', va = 'center', fontsize = 20)


# Status dos Pais por Nota Média Geral
plt.subplot(1, 2, 1)
sns.boxenplot(x='Pstatus', y='média', data=df, palette = 'Blues')
plt.xlabel('Status dos Pais')
plt.ylabel('Média Geral')
plt.title('Status dos Pais por Nota Média Geral', fontsize = 14)

# Status dos Pais por Consumo de Álcool em Dias Úteis
plt.subplot(1, 2, 2)
sns.violinplot(x='Pstatus', y='Dalc', data=df, palette = 'light:g')
plt.xlabel('Status dos Pais')
plt.ylabel('Consumo Álcool em Dias Úteis')
plt.title('Status dos Pais por Consumo de Álcool em Dias Úteis', fontsize = 14)


plt.show()
```

#### **A qualidade do relacionamento do aluno com seus familiares influencia nas notas e no consumo de álcool?**

```python
sns.set(style='whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
fig.suptitle('Análise por Qualidade dos Relacionamentos Familiares', ha = 'center', va = 'center', fontsize = 20)


# Qualidade por Nota Média Geral
y = df['média'].sort_values(ascending=False)
x = df['famrel']
plt.subplot(1, 2, 1)
sns.swarmplot(x, y, palette = 'Blues')
plt.xlabel('Qualidade dos Relacionamentos Familiares')
plt.ylabel('Média Geral')
plt.title('Qualidade por Nota Média Geral', fontsize = 14)

# Qualidade por Consumo de Álcool em Dias Úteis
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='famrel', y='Dalc', palette = 'light:g')
plt.xlabel('Qualidade dos Relacionamentos Familiares')
plt.ylabel('Consumo Álcool em Dias Úteis')
plt.title('Qualidade por Consumo de Álcool em Dias Úteis', fontsize = 14)


plt.show()
```

#### **Namoro influencia em notas?**

```python
df_v = df.copy().replace('no', ' Não namora').replace('yes', 'Namora').replace('F', 'Feminino ').replace('M', ' Masculino  ')
# 1
romantic_count = df_v.groupby('romantic')['média'].count().reset_index().rename(columns={'média': 'estudantes'})
# 2
romantic_sex_count = df_v.groupby(['romantic', 'sex'])[['média']].count().reset_index().sort_values('sex').rename(columns={'média': 'estudantes'})
# 3
df_v['romantic'] = df_v['romantic'].replace('no', ' Não namora').replace('yes', 'Namora')
score_romantic_mean = df_v.groupby(['romantic', 'sex'])[['média']].mean().reset_index().sort_values('romantic')
# 4
romantic_abstence = df_v.groupby(['romantic', 'sex'])[['absences']].mean().reset_index().sort_values('sex')
# 5
romantic_alcohol = df_v.groupby(['romantic', 'sex'])[['Dalc', 'Walc']].mean().reset_index().sort_values('sex')
v = df['Dalc'].mean()
```

`[Gráfico 1]`: Nº de Estudantes que namoram ou não

```python
fig = px.pie(romantic_count, values='estudantes', names='romantic', color='romantic',
             title='Estudantes que namoram ou não',)
fig.update_traces(textfont_size=14, textposition='inside', textinfo='percent+label+value')
fig.show()
```

---
`[Gráfico 2]`: Nº de Estudantes que namoram ou não por Sexo

```python
fig2 = px.bar(romantic_sex_count, x='romantic', y='estudantes', color='sex',
             title='Nº de Estudantes que namoram ou não por Sexo',
             text = 'estudantes',
)
fig2.update_layout( yaxis_title = 'Nº de estudantes', barmode='group',)
fig2.update_traces(textfont_size=13, textangle=0, cliponaxis=False)

fig2.show()
```

---
`[Gráfico 3]`: Desempenho de Estudantes que namoram ou não por Sexo

```python
fig3 = go.Figure(data=[
    go.Bar(name = 'Masculino',
           x = score_romantic_mean['romantic'].unique(),
           y = round(score_romantic_mean[score_romantic_mean['sex'] == ' Masculino  ']['média'], 2),),
    go.Bar(name = 'Feminino',
           x = score_romantic_mean['romantic'].unique(),
           y = round(score_romantic_mean[score_romantic_mean['sex'] == 'Feminino ']['média'], 2))
])

fig3.add_hline(y = df['média'].mean(), line_dash = 'dot',
              annotation_text = f'Valor Médio = {round(df["média"].mean(), 2)}', 
              annotation_position = 'top right',
              annotation_font_size = 13,
              line_width = 2.5, 
              )

fig3.update_layout(
    title = 'Desempenho de Estudantes que namoram ou não por Sexo',
    yaxis_title = 'Notas (0 - 20)',
    barmode='group',
)

fig3.show()
```
