```python
def calcular_media(row):
  return round((row['G1'] + row['G2'] + row['G3']) / 3, 2)

df['m√©dia'] = df.apply(lambda row: calcular_media(row), axis=1)

df.head()
```

## **Gr√°ficos com Matplotlib**
#### **Qual a propor√ß√£o dos estudantes que querem fazer faculdade?**


```python
dataT = df['higher'].value_counts(normalize=True).values
dataM = df[df['sex'] == 'M']['higher'].value_counts(normalize=True)
dataF = df[df['sex'] == 'F']['higher'].value_counts(normalize=True)

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Pretendem fazer Faculdade', ha = 'center', va = 'center', fontsize = 20)

axs[0].set_title('Todos')
axs[0].pie(
  dataT,
  labels=['Sim', 'N√£o'],
  shadow=True,  
  wedgeprops={'linewidth': 1, 'edgecolor': "blue"},
  explode=(0, 0.1),             
  autopct='%1.1f%%')

axs[1].set_title('Homens')
axs[1].pie(
  dataM,
  labels=['Sim', 'N√£o'],
  shadow=True,  
  wedgeprops={'linewidth': 1, 'edgecolor': "blue"},
  explode=(0, 0.1),
  autopct='%1.1f%%')

axs[2].set_title('Mulheres')
axs[2].pie(
  dataF,
  labels=['Sim', 'N√£o'],
  shadow=True,
  wedgeprops={'linewidth': 1, 'edgecolor': "blue"},
  explode=(0, 0.1),
  autopct='%1.1f%%')

plt.show()
```

#### **Querer fazer faculdade influencia no desempenho dos estudantes?**

```python
dataS = df[df['higher'] == 'yes']['m√©dia']
dataN = df[df['higher'] == 'no']['m√©dia']

dataMY = df[df['higher'] == 'yes'][df['sex'] == 'M']['m√©dia']
dataMN = df[df['higher'] == 'no'][df['sex'] == 'M']['m√©dia']

dataFY = df[df['higher'] == 'yes'][df['sex'] == 'F']['m√©dia']
dataFN = df[df['higher'] == 'no'][df['sex'] == 'F']['m√©dia']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Influ√™ncia do desejo de ir para Faculdade', ha = 'center', va = 'center', fontsize = 20)

axs[0].set_title('Todos')
axs[0].scatter(dataS.index, dataS, label='Desejam ir', color='blue')
axs[0].scatter(dataN.index, dataN, label='N√£o desejam ir', color='red')

axs[1].set_title('Homens')
axs[1].scatter(dataMY.index, dataMY, label='Desejam ir', color='blue')
axs[1].scatter(dataMN.index, dataMN, label='N√£o desejam ir', color='red')

axs[2].set_title('Mulheres')
axs[2].scatter(dataFY.index, dataFY, label='Desejam ir', color='blue')
axs[2].scatter(dataFN.index, dataFN, label='N√£o desejam ir', color='red')

plt.legend()
plt.show()
```

#### **Aulas pagas de refor√ßo influenciam no desempenho dos estudantes?**

```python
dataS = df[df['schoolsup'] == 'yes']['m√©dia']
dataN = df[df['schoolsup'] == 'no']['m√©dia']

dataMY = df[df['schoolsup'] == 'yes'][df['sex'] == 'M']['m√©dia']
dataMN = df[df['schoolsup'] == 'no'][df['sex'] == 'M']['m√©dia']

dataFY = df[df['schoolsup'] == 'yes'][df['sex'] == 'F']['m√©dia']
dataFN = df[df['schoolsup'] == 'no'][df['sex'] == 'F']['m√©dia']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Influ√™ncia do Refor√ßo', ha = 'center', va = 'center', fontsize = 20)

axs[0].set_title('Todos')
axs[0].scatter(dataS.index, dataS, label='Com refor√ßo', color='red')
axs[0].scatter(dataN.index, dataN, label='Sem refor√ßo', color='blue')

axs[1].set_title('Homens')
axs[1].scatter(dataMY.index, dataMY, label='Com refor√ßo', color='red')
axs[1].scatter(dataMN.index, dataMN, label='Sem refor√ßo', color='blue')

axs[2].set_title('Mulheres')
axs[2].scatter(dataFY.index, dataFY, label='Com refor√ßo', color='red')
axs[2].scatter(dataFN.index, dataFN, label='Sem refor√ßo', color='blue')

plt.legend()
plt.show()
```

## **Gr√°ficos com Seaborn**

```python
df['traveltime'].replace(1, '<15 min.', inplace = True)
df['traveltime'].replace(2, '15 a 30 min.', inplace = True)
df['traveltime'].replace(3, '30 min. a 1 hora', inplace = True)
df['traveltime'].replace(4, '>1 hora', inplace = True)
df['Pstatus'].replace('T', 'Pais Juntos', inplace = True)
df['Pstatus'].replace('A', 'Pais Divorciados', inplace = True)
df['famrel'].replace(1, 'Muito Ruim', inplace = True)
df['famrel'].replace(2, 'Ruim', inplace = True)
df['famrel'].replace(3, 'Neutro', inplace = True)
df['famrel'].replace(4, 'Bom', inplace = True)
df['famrel'].replace(5, 'Excelente', inplace = True)
```

#### **O tempo de deslocamento de casa para a escola influencia nas notas?**

```python
df_traveltime1 = df.pivot_table(values = 'm√©dia', index = 'traveltime', aggfunc = 'mean').sort_values(by = 'm√©dia', ascending = False).reset_index()


sns.set(style='whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
fig.suptitle('An√°lise por Tempo de Deslocamento', ha = 'center', va = 'center', fontsize = 20)

# Tempo de Deslocamento
plt.subplot(1, 2, 1)
sns.histplot(df.traveltime, palette = 'Blues') 
plt.xlabel('Tempo de Deslocamento')
plt.title('Distribui√ß√£o Tempo de Deslocamento', fontsize = 14)


# Tempo de Deslocamento por Nota M√©dia Geral
y = df_traveltime1['m√©dia'].sort_values(ascending=False)
x = df_traveltime1['traveltime']
plt.subplot(1, 2, 2)
sns.barplot(x, y, palette = 'Blues')
plt.xlabel('Tempo de Deslocamento')
plt.ylabel('M√©dia Geral')
plt.title('Tempo de Deslocamento por Nota M√©dia Geral', fontsize = 14)


plt.show()
```

#### **A qualidade do relacionamento do aluno com seus familiares influencia nas notas e no consumo de √°lcool?**

#### **Namoro influencia em notas?**

```python
df_v = df.copy().replace('no', ' N√£o namora').replace('yes', 'Namora').replace('F', 'Feminino ').replace('M', ' Masculino  ')
# 1
romantic_count = df_v.groupby('romantic')['m√©dia'].count().reset_index().rename(columns={'m√©dia': 'estudantes'})
# 2
romantic_sex_count = df_v.groupby(['romantic', 'sex'])[['m√©dia']].count().reset_index().sort_values('sex').rename(columns={'m√©dia': 'estudantes'})
# 3
df_v['romantic'] = df_v['romantic'].replace('no', ' N√£o namora').replace('yes', 'Namora')
score_romantic_mean = df_v.groupby(['romantic', 'sex'])[['m√©dia']].mean().reset_index().sort_values('romantic')
# 4
romantic_abstence = df_v.groupby(['romantic', 'sex'])[['absences']].mean().reset_index().sort_values('sex')
# 5
romantic_alcohol = df_v.groupby(['romantic', 'sex'])[['Dalc', 'Walc']].mean().reset_index().sort_values('sex')
v = df['Dalc'].mean()
```

`[Gr√°fico 1]`: N¬∫ de Estudantes que namoram ou n√£o

```python
fig = px.pie(romantic_count, values='estudantes', names='romantic', color='romantic',
             title='Estudantes que namoram ou n√£o',)
fig.update_traces(textfont_size=14, textposition='inside', textinfo='percent+label+value')
fig.show()
```

---
`[Gr√°fico 2]`: N¬∫ de Estudantes que namoram ou n√£o por Sexo

```python
fig2 = px.bar(romantic_sex_count, x='romantic', y='estudantes', color='sex',
             title='N¬∫ de Estudantes que namoram ou n√£o por Sexo',
             text = 'estudantes',
)
fig2.update_layout( yaxis_title = 'N¬∫ de estudantes', barmode='group',)
fig2.update_traces(textfont_size=13, textangle=0, cliponaxis=False)

fig2.show()
```

---
`[Gr√°fico 3]`: Desempenho de Estudantes que namoram ou n√£o por Sexo

```python
fig3 = go.Figure(data=[
    go.Bar(name = 'Masculino',
           x = score_romantic_mean['romantic'].unique(),
           y = round(score_romantic_mean[score_romantic_mean['sex'] == ' Masculino  ']['m√©dia'], 2),),
    go.Bar(name = 'Feminino',
           x = score_romantic_mean['romantic'].unique(),
           y = round(score_romantic_mean[score_romantic_mean['sex'] == 'Feminino ']['m√©dia'], 2))
])

fig3.add_hline(y = df['m√©dia'].mean(), line_dash = 'dot',
              annotation_text = f'Valor M√©dio = {round(df["m√©dia"].mean(), 2)}', 
              annotation_position = 'top right',
              annotation_font_size = 13,
              line_width = 2.5, 
              )

fig3.update_layout(
    title = 'Desempenho de Estudantes que namoram ou n√£o por Sexo',
    yaxis_title = 'Notas (0 - 20)',
    barmode='group',
)

fig3.show()
```
